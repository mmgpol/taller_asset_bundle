```python
from pyspark.sql import functions as F
```

```python
# Transform (join + m√©tricas)
df_sales = (
    df_orders
      .join(df_customers, on="customer_id", how="left")
      .join(df_products,  on="product_id",  how="left")
      .withColumn("revenue", F.col("qty") * F.col("unit_price"))
      .select(
          "order_id", "order_date",
          "customer_id", "name", "country", "segment",
          "product_id", "product_name", "category",
          "qty", "unit_price", "revenue"
      )
)

# Write Silver
(df_sales.write.mode("overwrite").format("delta").save(f"{silver_root}/sales"))

display(df_sales.orderBy("order_date", "order_id"))
print("Silver escrito en:", f"{silver_root}/sales"
```
